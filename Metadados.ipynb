```python
def AjustaMetadados(Dataframe):

    train = ABT
    #verifica os tipos de variáveis presentes na tabela
    t = []
    for i in train.columns:
        t.append(train[i].dtype)
        
    n = []
    for i in train.columns:
        n.append(i)
        
    aux_t = pd.DataFrame(data=t, columns=["Tipos"])
    aux_n = pd.DataFrame(data=n, columns=["Features"])
    
    df_tipovars = pd.merge(aux_n, aux_t, how='left', left_index=True, right_index=True)
    
    # role
    data=[]
    for f in train.columns:
        
        if f =='TARGET':
            role = 'target'
        elif f == 'SK_ID_CURR':
            role = 'id'
        else:
            role = 'input'
            
        if f == 'TARGET':
            level = 'binary'
        elif train[f].dtype == 'object' or f == 'SK_ID_CURR':
            level = 'nominal'
        elif train[f].dtype in ['float','float64']:
            level = 'interval'
        elif train[f].dtype in ['int', 'int64']:
            level = 'ordinal'
        else:
            level = 'NA'
        
        keep = True
        if f == 'SK_ID_CURR':
            keep = False
            
        dtype = train[f].dtype
        
        f_dict = {
            'Features': f,
            'Role': role,
            'Level': level,
            'Keep': keep,
            'Tipo': dtype
        }
        data.append(f_dict)
        
    meta = pd.DataFrame(data, columns=['Features', 'Role', 'Level', 'Keep', 'Tipo'])
    
    #Cardinalidade
    card = []
    
    v = train.columns
    for f in v:
        dist_values = train[f].value_counts().shape[0]
        f_dict = {
            'Features': f,
            'Cardinality': dist_values
            }
        card.append(f_dict)
    
    card = pd.DataFrame(card, columns=['Features', 'Cardinality'])
    
    metadados_train = pd.merge(meta, card, on='Features')
    
    return metadados_train
    
```

```python
metadados = AjustaMetadados(ABT)
metadados
```

### Salvando os metadados

```python
metadados.to_csv("metadados.csv")
```
